Heatmap:-Heatmap is a graphical way to visualize visitor behavior data in the form of hot and cold spots employing a warm-to-cool color scheme

Histogram:-It encapsulates discrete or continuous data that is measured on an interval scale. It frequently serves as an example of the main characteristics of the data distribution in an easy-to-understand format.

Bar Plot:-Bar graph (bar chart) is a graph that represents the categorical data using rectangular bars. The bar graph shows the comparison between discrete categories.

CatPlot (Categorical Plot):-This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations.

Density Plot:-Density Plot is the continuous and smoothed version of the Histogram estimated from the data

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
tita=sns.load_dataset('titanic')
dia=sns.load_dataset("diamonds")
iris=sns.load_dataset('iris')
flights=sns.load_dataset('flights')
```

```python
tita
```

```python
sns.lineplot(x =fmri.timepoint, y = fmri.signal ,hue=fmri.event,style=fmri.region)
```

```python
sns.boxplot(x=tita.pclass,y=tita.age)
```

```python
sns.boxplot(x=tita.pclass,y=tita.fare)
```

```python
sns.histplot(x=dia.price,hue=dia.cut)
```

```python
sns.pairplot(iris,hue='species')
```

```python
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```
